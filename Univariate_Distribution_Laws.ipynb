All Distribution Laws in Python, using the scipy.stats library. Then I will create these distributions law directly from Numpy (starting from a uniform distribution and transforming it)



Binomial Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

Uniform Distribution

```python
from scipy.stats import uniform
# Parameters
a = 0 # lower bound
b = 10 # upper bound
# Creating a Uniform distribution
x = np.linspace(a, b, 100)
pdf = uniform.pdf(x, loc=a, scale=b-a)
cdf = uniform.cdf(x, loc=a, scale=b-a)

# Plotting
plt.plot(x, pdf)
plt.title('Uniform Distribution PDF')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()

plt.plot(x, cdf)
plt.title('Uniform Distribution CDF')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

```python
from scipy.stats import binom
# Parameters
n = 20 # number of trials
p = 0.6 # probability of success

# Creating a binomial distribution
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)
cdf = binom.cdf(x,n,p)

# Plotting
plt.bar(x, pmf)
plt.title('Binomial Distribution PMF')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()

plt.bar(x, cdf)
plt.title('Binomial Distribution CDF')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```

Poisson Distribution


```python
from scipy.stats import poisson

# Parameters
lambda_poisson = 2 # average rate (events per interval)

# Creating a Poisson distribution
x = np.arange(0, 20)
pmf = poisson.pmf(x, lambda_poisson)
cdf = poisson.cdf(x, lambda_poisson)

# Plotting
plt.bar(x, pmf)
plt.title('Poisson Distribution PMF')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()

plt.bar(x, cdf)
plt.title('Poisson Distribution CDF')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()

```

GEOMETRIC DISTRIBUTION
The Geometric distribution models the number of trials needed to get the first success
in a sequence of independent Bernoulli trials, each with the same probability of
success.

```python
from scipy.stats import geom
# Parameters
p = 0.3 # probability of success

# Creating a Geometric distribution
x = np.arange(1, 11)
pmf = geom.pmf(x, p)
cdf = geom.cdf(x, p)
# Plotting
plt.bar(x, pmf)
plt.title('Geometric Distribution PMF')
plt.xlabel('Number of Trials')
plt.ylabel('Probability')
plt.show()

plt.bar(x, cdf)
plt.title('Geometric Distribution CDF')
plt.xlabel('Number of Trials')
plt.ylabel('Probability')
plt.show()
```

Normal Distribution


```python
from scipy.stats import norm
# Parameters
mu = 0 # mean
sigma = 1 # standard deviation

# Creating a Normal distribution
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
pdf = norm.pdf(x, mu, sigma)
cdf = norm.cdf(x, mu, sigma)

# Plotting
plt.plot(x, pdf)
plt.title('Normal Distribution PDF')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()

plt.plot(x, cdf)
plt.title('Normal Distribution CDF')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

Exponential Distribution

```python
from scipy.stats import expon
# Parameters
lambda_ = 1 # rate parameter (lambda)

# Creating an Exponential distribution
x = np.linspace(0, 10, 100)
pdf = expon.pdf(x, scale=1/lambda_)
cdf = expon.cdf(x, scale=1/lambda_)

# Plotting
plt.plot(x, pdf)
plt.title('Exponential Distribution PDF')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()


plt.plot(x, cdf)
plt.title('Exponential Distribution CDF')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```
